```python
data_loader=Data_load(input_line=64, frames=5)
```

```python
lidar_frames,gt_frames,img_frames=data_loader.read_frames(batch=1, if_removal=False)
```

```python
feature_params = dict( maxCorners = 100,
                       qualityLevel = 0.4,
                       minDistance = 7,
                       blockSize = 7,
                       useHarrisDetector=False)
lk_params = dict( winSize  = (15,15),
                  maxLevel = 2,
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
```

```python
gray_image = cv2.cvtColor(img_frames[0,:,:,:], cv2.COLOR_BGR2GRAY)
mask_img = np.asarray(lidar_frames[0,:,:,0],dtype=np.uint8)
features = cv2.goodFeaturesToTrack(gray_image, mask=mask_img, **feature_params)
# orb = cv2.ORB_create()
# features = np.asarray(orb.detect(gray_image,mask_img))
print(features.shape)
print(lidar_frames.shape)
print(features)
print(features[0,0,1])
print(lidar_frames[0,152,394,0])
print(len(features))
```

```python
next_gray_img = cv2.cvtColor(img_frames[1,:,:,:],cv2.COLOR_BGR2GRAY)
next_features,status,error = cv2.calcOpticalFlowPyrLK(gray_image,next_gray_img, features,None,**lk_params)
```

```python
good_old = features[status==1]
good_new = next_features[status==1]
color = np.random.randint(0,255,(100,3))
for i,(new,old) in enumerate(zip(good_new,good_old)) :
    a,b = new.ravel()
    c,d = old.ravel()
    frame = cv2.circle(img_frames[0,:,:,:],(a,b),5,color[i].tolist(),10)
    next_frame = cv2.circle(img_frames[1,:,:,:],(c,d),5,color[i].tolist(),10)
```

```python
plt.imshow(frame)
plt.show()

plt.imshow(next_frame)
plt.show()
```
