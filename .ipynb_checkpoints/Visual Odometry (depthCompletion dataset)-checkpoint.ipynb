```python
from data_read import *
from odometry_functions_depthcompletion_dataset import *
import matplotlib.pyplot as plt
import numpy as np
import cv2
import torch
from scipy.optimize import least_squares
import os
import math
import matplotlib.pyplot as plt
import time
```

```python
feature_params = dict( maxCorners = 100,
                       qualityLevel = 0.3,
                       minDistance = 7,
                       blockSize = 7,
                       useHarrisDetector=False)
lk_params = dict( winSize  = (15,15),
                  maxLevel = 6,
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

num_frames = 5
n_features = 45
data_loader=Data_load(input_line=16, frames=num_frames)
lidar_frames,gt_frames,img_frames,index_list=data_loader.read_frames(batch_size=1, if_removal=False, index = 36695 )
```

```python
frame_id = 0

# scene=data_loader.img_path[index_list[0]].split('/')[5].split('_')[:3]
# scene_date = scene[0]
# camera_id = data_loader.img_path[index_list[0]].split('/')[-3]
# for i in range(1,len(scene)) : 
#     scene_date += '_' + scene[i]
# K = get_calib_file(scene_date,camera_id) # eg: 2011_09_26, image_03
K = data_loader.get_camera_matrix()

T_list = []

while (frame_id < num_frames-1):
    features = find_keypoints(img_frames[frame_id,:,:,:],lidar_frames[frame_id,:,:,0],feature_params,n_features)
    next_features, status, error = track_features(img_frames[frame_id,:,:,:],features,img_frames[frame_id+1,:,:,:],lk_params)
    features = features[status==1]
    next_features = next_features[status==1]

    image = np.copy(img_frames[frame_id,:,:,:])
    next_image = np.copy(img_frames[frame_id+1,:,:,:])
    for i in range(len(features)):
        cv2.circle(image,(features[i][0],features[i][1]),10,(0,0,255),-1)
        cv2.circle(next_image,(next_features[i][0],next_features[i][1]),10,(0,0,255),-1)

    plt.imshow(image)
    plt.show()

    plt.imshow(next_image)
    plt.show()

    pts_3d_1 = compute_3D_points(lidar_frames[frame_id,:,:,:],features,K)
    pts_2d_2 = []
    for i in range(len(features)):
        x = int(next_features[i,0])
        y = int(next_features[i,1])
        pts_2d_2.append([x,y])
    pts_2d_2 = np.asarray(pts_2d_2)
    print("Total points : ",len(pts_2d_2))
    P = np.hstack((K,np.zeros((3,1))))

    pts_3d_1 = np.expand_dims(pts_3d_1.astype(np.float32),axis=1)
    pts_2d_2 = np.expand_dims(pts_2d_2.astype(np.float32),axis=1)
    ret = cv2.solvePnPRansac(pts_3d_1,pts_2d_2,P[:3,:3],distCoeffs=None)

    print("Solution found : ",ret[0])

#     r = genEulerZXZMatrix(-ret[1][0][0],-ret[1][1][0],-ret[1][2][0])
#     transf_matrix = np.hstack((r,-ret[2]))
#     transf_matrix = np.vstack((transf_matrix,np.asarray([0,0,0,1])))

    r, _ = cv2.Rodrigues(ret[1])
    transf_matrix = np.hstack((r,ret[2]))
    transf_matrix = np.vstack((transf_matrix,np.array([0,0,0,1])))
    print("T (from solvePnPRansac) : ")
    print(transf_matrix)
    T_list.append(transf_matrix)
    
    pts_3d_1 = np.hstack( (np.squeeze(pts_3d_1), np.ones(( len(pts_3d_1),1) )))
    pts_2d_2 = np.hstack( (np.squeeze(pts_2d_2), np.ones(( len(pts_2d_2),1) )))

    dSeed = np.zeros(len(pts_3d_1))
    opt = least_squares(func,dSeed,method='lm',verbose=0,args=(pts_3d_1,pts_2d_2,P))
    R = genEulerZXZMatrix(opt.x[0],opt.x[1],opt.x[2])
    t = np.asarray([[opt.x[3]],[opt.x[4]],[opt.x[5]]])
    T = np.hstack((R,t))
    T = np.vstack((T,np.asarray([[0,0,0,1]])))
    print("T (from scipy least_squares) : ")
    print(T)
    frame_id += 1
```

```python
frame_id = 0
pts3_tempList = []
nc = npCloud(K,lidar_frames[frame_id,:,:,0].shape[0],lidar_frames[frame_id,:,:,0].shape[1])
# finding 3D points for all valid pixels in depth image
while (frame_id < num_frames) :
    depth = lidar_frames[frame_id,:,:,0]
    pts3 = nc.calc3DPoints(depth)
    pts3_tempList.append(pts3)
    frame_id += 1
    
# finding transformation from ith frame to the last frame
i = len(T_list)-1
T_intermToLast = [T_list[i]]
while (i>0) :
    T_temp = np.matmul(T_intermToLast[0],T_list[i-1])
    T_intermToLast.insert(0,T_temp)
    i -= 1

# representing all 3D points wrt the last frame
frame_id = 0
pts3_concatenated = np.asarray([]).reshape(3,0)
while (frame_id < num_frames-1): 
    pts3 = pts3_tempList[frame_id]
    pts3 = np.vstack((pts3,np.ones((1,pts3.shape[1])))) # converting to homogeneous coordinates
    T = T_intermToLast[frame_id]
    pts3 = np.matmul(T,pts3)
    pts3_concatenated = np.hstack((pts3_concatenated,pts3[:3,:]))
    frame_id += 1

pts3_concatenated = np.hstack((pts3_concatenated,pts3_tempList[num_frames-1][:3,:]))
depth_concatenated = np.zeros(lidar_frames[0,:,:,0].shape)

# finding 2D projections of all 3D points in the last frame
pts2 = np.matmul(K,pts3_concatenated)
pts2[:2,:] /= pts2[2,:]
u = pts2[0,:].astype(np.int)
v = pts2[1,:].astype(np.int)
z = pts2[2,:]

print(num_frames," sparse lidar frames ")
for i in range(num_frames):
    print("Id : ",i)
    plt.imshow(lidar_frames[i,:,:,0])
    plt.show()

print("Num points after concatenation : ",u.shape)
u,v,z = nc.checkLimits(u,v,z)
# idx = np.where(u<nc.width)
# u = u[idx]
# v = v[idx]
# z = z[idx]

# idx = np.where(v<nc.height)
# u = u[idx]
# v = v[idx]
# z = z[idx]

# idx = np.where(u>=0)
# u = u[idx]
# v = v[idx]
# z = z[idx]

# idx = np.where(v>=0)
# u = u[idx]
# v = v[idx]
# z = z[idx]

# idx = np.where(z>=0)
# u = u[idx]
# v = v[idx]
# z = z[idx]

print("Num points after concatenation that fall on the image plane : ",u.shape)
depth_concatenated[v,u] = z

print("Ground truth depth image : ")
plt.imshow(gt_frames[num_frames-1,:,:,0])
plt.show()

print("Concatenated depth image : ")
plt.imshow(depth_concatenated)
plt.show()

gt_mask = np.where(gt_frames[num_frames-1,:,:,0]!=0,1,0)
depth_mask = np.where(depth_concatenated!=0,1,0)
depth_gt_diff = gt_mask * depth_mask * (depth_concatenated - gt_frames[num_frames-1,:,:,0])

depth_gt_diff *= depth_gt_diff
print("RMSE between GT & Concatenated depth image : ",1000*np.sqrt(np.mean(depth_gt_diff)))
print("--------")

depth_concatenated = np.where(depth_concatenated>0,100-depth_concatenated,0)
depth_concatenated_dilated = cv2.dilate(depth_concatenated,np.ones((5,5),np.uint8))
depth_concatenated_dilated = np.where(depth_concatenated_dilated>0,100-depth_concatenated_dilated,0)
depth_mask = np.where(depth_concatenated_dilated!=0,1,0)
depth_gt_diff = gt_mask * depth_mask * (depth_concatenated_dilated - gt_frames[num_frames-1,:,:,0])

print("Dilated depth concatenated image : ")
plt.imshow(depth_concatenated_dilated)
plt.show()

depth_gt_diff *= depth_gt_diff
print("RMSE between GT & Dilated Depth concatenated image : ", 1000*np.sqrt(np.mean(depth_gt_diff)))
print("--------")


depth_dilated = np.where(lidar_frames[num_frames-1,:,:,0]>0,100-lidar_frames[num_frames-1,:,:,0],0)
depth_dilated = cv2.dilate(depth_dilated,np.ones((5,5),np.uint8))
depth_dilated = np.where(depth_dilated>0,100-depth_dilated,0)
depth_mask = np.where(depth_dilated!=0,1,0)
depth_gt_diff = gt_mask * depth_mask * (depth_dilated - gt_frames[num_frames-1,:,:,0])

print("Dilated sparse depth image : ")
plt.imshow(depth_dilated)
plt.show()

depth_gt_diff *= depth_gt_diff
print("RMSE between GT & Dilated sparse depth image : ",1000*np.sqrt(np.mean(depth_gt_diff)))
```

```python
from slic import SuperPixels
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io, color
from sklearn.neighbors import KDTree

spx_list = []
min_diff = 1
K = data_loader.get_camera_matrix() # eg: 2011_09_26, image_03
nc = npCloud(K,lidar_frames[0].shape[0],lidar_frames[0].shape[1])
outlier_indices = {}
count_rejected_points = 0

for frame_id in range(num_frames-1) :
    img1 = img_frames[frame_id,:,:,:]
    lidar_img1 = lidar_frames[frame_id,:,:,0]
    
    img2 = img_frames[frame_id+1,:,:,:]
    lidar_img2 = lidar_frames[frame_id+1,:,:,0]
    
    spx1 = SuperPixels(img1,lidar_img1)
    spx2 = SuperPixels(img2,lidar_img2)
    
    numSegments = 300
    compactness = 100.0
    segments11 = slic(img_frames[frame_id,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
    segments21 = slic(img_frames[frame_id+1,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
    
    numSegments = 25
    compactness = 10.0
    segments12 = slic(img_frames[frame_id,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
    segments22 = slic(img_frames[frame_id+1,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
    
    spx1.init_clusters(segments11)
    spx1.init_clusters(segments12)
    
    spx2.init_clusters(segments21)
    spx2.init_clusters(segments22)
    
    cluster_centres = []
    for cluster_id in spx1.clusters :
        cluster_centres.append([spx1.clusters[cluster_id].mean_col, spx1.clusters[cluster_id].mean_row]) # (x,y)
    cluster_centres = np.asarray(cluster_centres)
    spx1.update_KDtree(KDTree(cluster_centres,leaf_size=2))
    
    for candidate_index in spx2.clusters : 
#         print(candidate_index)
        if  spx2.clusters[candidate_index].num_lidar_points!= 0 :
            d_list = []
    #         print("candidate index : {}".format(candidate_index))
#             plt.imshow(color.lab2rgb(spx2.clusters[candidate_index].masked_img))
#             plt.show()
            dist, nearest_cluster_indices = spx1.tree.query(spx2.clusters[candidate_index].get_centre_xy(),k=5)
#             print("dist ",dist)
#             print("nearest cluster ", nearest_cluster_indices )
            Dmin, min_cluster_index = -1, -1
            for j,i in enumerate(nearest_cluster_indices[0]) :
                d_list.append(spx2.clusters[candidate_index].get_distance(spx1.clusters[i],dist[0][j],0.9))
    #             print("d-total : ",d_list[-1]," ds : ",dist[0][j])
            d_list = np.asarray(d_list)
            cluster_match_index = nearest_cluster_indices[0][np.argmin(d_list)]
#             plt.imshow(color.lab2rgb(spx1.clusters[cluster_match_index].masked_img))
#             plt.show()
            if (spx1.clusters[cluster_match_index].num_lidar_points!=0) :
                spx1.clusters[cluster_match_index].match_index_next_img = candidate_index
                spx2.clusters[candidate_index].match_index_prev_img = cluster_match_index
    
    T = T_list[frame_id]
    depth_concatenated_temp = np.zeros(spx1.lidar_img.shape)
    outlier_indices[frame_id] = []
    for index1 in spx1.clusters : 
        index2 = spx1.clusters[index1].match_index_next_img
        if index2 == -1 :
            continue
        pts3d = nc.calc3DPoints(spx1.clusters[index1].masked_lidar_img)
        pts_transformed = nc.transform_points(pts3d,T)
        d1 = np.sum(pts_transformed[2,:])/pts_transformed.shape[1]
        d2 = spx2.clusters[index2].mean_depth
        if (abs(d1-d2) > min_diff) : 
            outlier_indices[frame_id].append(index1)  
            count_rejected_points += spx1.clusters[index1].num_lidar_points
        else : 
            pts3d = nc.calc3DPoints(spx1.clusters[index1].masked_lidar_img)
            pts3d_transformed = nc.transform_points(pts3d,T) # homogeneous coordinates
            u,v,z = nc.calc2DPoints(pts3d_transformed[:3,:]) 
            depth_concatenated_temp[v,u] = z

            pts3d_current = nc.calc3DPoints(spx2.clusters[index2].masked_lidar_img)
            u,v,z = nc.calc2DPoints(pts3d_current)
            depth_concatenated_temp[v,u] = z
    
    lidar_frames[frame_id+1,:,:,0] = depth_concatenated_temp
    
print("done")
# for frame_id in range(num_frames-1) :  
#     numSegments = 300
#     compactness = 100.0
#     segments11 = slic(img_frames[frame_id,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
#     segments21 = slic(img_frames[frame_id+1,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)

#     numSegments = 25
#     compactness = 10.0
#     segments12 = slic(img_frames[frame_id,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
#     segments22 = slic(img_frames[frame_id+1,:,:,:],n_segments = numSegments, sigma = 5, compactness=compactness, max_iter=100, convert2lab=True)
    
#     if frame_id == 0 :
#         spx_list[0].init_clusters(segments11)
#         spx_list[0].init_clusters(segments12)
#         spx_list[1].init_clusters(segments21)
#         spx_list[1].init_clusters(segments22)
        
# for spx in spx_list : 
#     print(len(spx.clusters))

# # fig = plt.figure("Superpixels %d segments, image : %d" % (numSegments,0))
# # ax = fig.add_subplot(1,1,1)
# plt.imshow(mark_boundaries(img1,segments1))
# plt.show()
# # fig2 = plt.figure("Superpixels %d segments, image : %d" % (numSegments,1))
# # ax2 = fig2.add_subplot(1,1,1)
# plt.imshow(mark_boundaries(img2,segments2))
# plt.show()
# # fig5 = plt.figure("Superpixels %d segments, image : %d" % (numSegments,0))
# # ax5 = fig5.add_subplot(1,1,1)
# plt.imshow(mark_boundaries(img1,segments3))
# plt.show()
# # fig6 = plt.figure("Superpixels %d segments, image : %d" % (numSegments,1))
# # ax6 = fig6.add_subplot(1,1,1)
# plt.imshow(mark_boundaries(img2,segments4))
# plt.show()
# # fig7 = plt.figure("Original image 0")
# # ax7 = fig7.add_subplot(1,1,1)
# plt.imshow(img1)
# plt.show()
# # fig8 = plt.figure("Original image 1")
# # ax8 = fig8.add_subplot(1,1,1)
# plt.imshow(img2)
# plt.show()
# # fig9 = plt.figure("Original image 2")
# # ax9 = fig9.add_subplot(1,1,1)
# plt.imshow(img_frames[2])
# plt.show()
# # fig10 = plt.figure("Lidar img")
# # ax10 = fig10.add_subplot(1,1,1)
# plt.imshow(lidar_frames[2,:,:,0])
# plt.show()

# spx1 = SuperPixels(img1,lidar_img1)
# spx1.init_clusters(segments1)
# spx1.init_clusters(segments3)

# spx2 = SuperPixels(img2,lidar_img2)
# spx2.init_clusters(segments2)
# spx2.init_clusters(segments4)
```
